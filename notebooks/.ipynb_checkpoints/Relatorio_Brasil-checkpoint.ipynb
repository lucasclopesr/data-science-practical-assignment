# COVID-19 no Brasil

Esse Notebook objetiva analisar o impacto das medidas preventivas no combate à COVID-19 nos estados brasileiros. Para tal, faz-se necessário vizualizar como a pandemia afetou cada Estado ao longo do tempo. Usaremos como variável de resposta a quantidade de casos confirmados.

## Dados 

A fonte de dados utilizada para essa análise é o CSV casos_acumulados. A coleta desses dados contempla o 1° dia em que cada estado relatou casos confirmados até o 190° dia (aproximadamente 12/10). 

Alguns campos dessa tabela encontravam-se vazios. Dessa forma, a limpeza dos dados consistiu no preenchimento desses campos pela média entre o dia anterior e o dia seguinte.

Segue abaixo a leitura dos dados (a limpeza dos dados foi feita externamente e, portanto, não é abordado aqui):

```python
df = pd.read_csv('../databases/casos_acumulados.csv', index_col='Dias desde o início da epidemia')
df
```

Para fins de uma melhor visualização em termos de casos ao longo do tempo, será apresentado abaixo os gráficos da quantidade de casos em todos os estados e em cada estado separadamente nos primeiros 30 dias, e, posteriormente, ao longo de 6 meses (aproximadamente o tempo em que a pandemia está perdurando até então).

## Casos ao longo dos primeiros 30 dias

```python
primeiro_mes = df['Dia 1':'Dia 30']
primeiro_mes.head()
```

```python
plt = primeiro_mes.plot(figsize=(10,10), title='Quantidade de casos por estado ao longo do primeiro mês')
plt.set(ylabel='Número de casos')
```

Esse gráfico representa a quantidade de casos acumulados em cada um dos estados ao longo de um mês (é utilizada a escala logarítmica). A princípio, é possível observar que alguns estados tem sua curva tendendo à um crescimento exponencial, enquanto outros mantiveram um crescimento tendendo ao linear. Segue abaixo a visualização individual da quantidade de casos por estado:

```python
plt = primeiro_mes.plot(subplots=True, 
              title='Quantidade de casos por estado ao longo dos primeiros 30 dias', 
              layout=(9,3), 
              figsize=(10,10),
              ylim=(0,3000))
```

Observando agora os estados de maneira individual, é possível observar que aqueles mais próximos do crescimento exponencial foram estados como o Ceará, Rio de janeiro e São Paulo, enquanto os estados com crecimento mais próximo do linear foram Acre, Alagoas, Tocantins e Acre. 

Por se tratar de muitos Estados, a análise de todos tornaria o trabalho maçante. Dessa forma, a análise será focada nesses estados citados acima e, portanto, nas medidas preventivas adotadas por eles.

Antes de serem abordadas o impacto das medidas preventivas, é necessário primeiro analisar o comportamento da curva desses mesmos estados ao longo de 6 meses. Segue abaixo:

---
## Casos ao longo de 6 meses

O gráfico abaixo mostra o comparativo de curvas de crescimento da _COVID19_ em cada estado desde o primeiro caso reportado naquele estado.

```python
plt = df.plot(figsize=(10,10), title='Quantidade de casos por estado ao longo dos primeiros 6 meses')
plt.set(ylabel='Número de casos')
```

De maneira geral, observa-se que a curva da maioria dos estados tende a se aproximar da linear, exceto pelo estado de São Paulo, que tem seu comportamento exponencial (a curva está mais acentuada do que a mostrada para os primeiros 30 dias). 

Podemos também visualizar de forma isolada:

```python
plt = df.plot(subplots=True, 
              title='Quantidade de casos por estado ao longo dos primeiros 6 meses', 
              layout=(9,3), 
              figsize=(10,10),
              ylim=(0,600000))
```

Observando agora os estados individualmente, nota-se que a curva de São Paulo foi a única que cresceu em relação a de 30 dias, enquanto no Rio de Janeiro e Ceará , decresceu. Em relação aos estados com crescimento próximo do linear (Alagoas, Tocantins e Acre), nota-se um aumento ao longo dos 6 meses, mas ainda com crescimento próximo do linear.

Após a visualização desses dados, torna-se possível a análise do impacto das medidas preventivas. 

Antes disso, faz-se necessário lembrar que existem variáveis que não serão abordadas de maneira direta nessa análise, como: quantidade de habitantes por estado, quantidade de riquesa do estado, dentre outros, por adicionar complexidade ao processo. Dessa forma, ao final, apesar das análises e das conclusões, perguntas que envolvam a possível influência dessas variáveis surgirão, o que pode se tornar pauta de análises mais completas no futuro.
